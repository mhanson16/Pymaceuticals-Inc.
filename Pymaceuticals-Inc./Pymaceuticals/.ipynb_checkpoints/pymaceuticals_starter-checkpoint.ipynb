### Analysis

- Capomulin and Ramicane were tested on the most number of mice at about 275 each, Propriva tested the least number of times at about 160, while all other regimens were tested relatively euqally at 175 mice treated.
- Gender tested came out failry with 51.7% male and 49.3% female. 
- Capomulin and Ramicane have conisistantly lower tumor sizes while both Infubinol and Ceftamin have a mean tumor size roughly 15 mm3 greater. Though, Infubinol had one outlier subject ending the treatment with a tumor size of about 36 mm3.
- In the first 20 days of testing with Capomulin there is a steady increase of about 3 mm3 in tumor size then a drastic shrinking by a total on 6 mm3 in the final 25 days, leaving a net loss of 3 mm3. 
- When testing Capomulin's effect on average tumor size and mouse weight we see a strong correlation between these factors supported by the scatter plot pattern and an r value of 0.84.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import math
import numpy as np

pd.set_option('display.max_rows', None)

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_data_complete = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
mouse_data_complete.head()
```

```python
# Checking the number of mice. 249
mice_count = mouse_data_complete['Mouse ID'].nunique()
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice_ID = mouse_data_complete[mouse_data_complete[['Mouse ID', 'Timepoint']].duplicated() == True]
duplicate_mice_ID = duplicate_mice_ID['Mouse ID'].unique()
duplicate_mice_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_ID_df = mouse_data_complete.loc[mouse_data_complete['Mouse ID'] == duplicate_mice_ID[0]]
duplicate_mice_ID_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_data_complete_df = mouse_data_complete.drop_duplicates(subset=['Mouse ID', 'Timepoint'])
clean_mouse_data_complete_df.head()
```

```python
# Checking the number of mice in the clean DataFrame. 248
mice_count_clean = clean_mouse_data_complete_df['Mouse ID'].nunique()
mice_count_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_tumor_vl = clean_mouse_data_complete_df.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
med_tumor_vl = clean_mouse_data_complete_df.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
tumor_vl_var = clean_mouse_data_complete_df.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
tumor_vl_std_dev = clean_mouse_data_complete_df.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
tumor_vl_std_err = clean_mouse_data_complete_df.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

regimen_summary = {'Mean Tumor Volume': mean_tumor_vl,
                   'Median Tumor Volume': med_tumor_vl,
                   'Tumor Volume Variance': tumor_vl_var,
                   'Tumor Volume Std. Dev.': tumor_vl_std_dev,
                   'Tumor Volume Std. Err.': tumor_vl_std_err}

regimen_summary = pd.DataFrame(regimen_summary)

# Display results
regimen_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
regimen_agg_summary = clean_mouse_data_complete_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean", "median", "var", "std", "sem"])
regimen_agg_summary
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

#capomulin_1509 = clean_mouse_data_complete_df.loc[clean_mouse_data_complete_df["Drug Regimen"] == "Capomulin" ]

capomulin_l509_df = clean_mouse_data_complete_df.loc[(clean_mouse_data_complete_df["Mouse ID"] == "l509") & (clean_mouse_data_complete_df["Drug Regimen"] == "Capomulin")]
capomulin_l509_df = capomulin_l509_df[["Timepoint","Tumor Volume (mm3)"]]
capomulin_l509_df = capomulin_l509_df.set_index(["Timepoint"])

capomulin_l509_tp = capomulin_l509_df.index.values
capomulin_l509_tumor_vl = capomulin_l509_df.values



plt.plot(capomulin_l509_tp, capomulin_l509_tumor_vl)
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

capomulin_avg_df = clean_mouse_data_complete_df.loc[(clean_mouse_data_complete_df["Drug Regimen"] == "Capomulin")]
capomulin_avg_df = capomulin_avg_df.groupby("Mouse ID")


capomulin_tumor_avg = capomulin_avg_df["Tumor Volume (mm3)"].mean()
capomulin_weight_avg = capomulin_avg_df["Weight (g)"].mean()

plt.scatter(capomulin_weight_avg, capomulin_tumor_avg)
plt.title("Capomulin Treatment")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(capomulin_tumor_avg,capomulin_weight_avg)
print(f"The correlation coefficient between mouse weight and the average tumor volume is {round(st.pearsonr(capomulin_tumor_avg,capomulin_weight_avg)[0],2)}")


cap_slope, cap_int, cap_r, vc_p, cap_std_err = st.linregress(capomulin_weight_avg, capomulin_tumor_avg)
cap_fit = cap_slope * capomulin_weight_avg + cap_int
#same as above
plt.scatter(capomulin_weight_avg, capomulin_tumor_avg)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

#trend line
plt.plot(capomulin_weight_avg,cap_fit,"--",color="red")


plt.show()

```
